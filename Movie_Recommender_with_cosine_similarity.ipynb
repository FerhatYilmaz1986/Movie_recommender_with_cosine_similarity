```python
import numpy as np
import pandas as pd
import os
```

```python
parent_directory = os.path.dirname(os.getcwd())
path = parent_directory+'\data\\'

# read the original data
data = pd.read_csv(path+'IMDB-MOVIE-DATA.csv', header=0)
```

```python
data.head()
```

```python
data.columns
```

```python
# columns to remove
columns_to_remove = ['Rank', 'Title', 'Description', 'Actors']

data_subset = data.drop(columns_to_remove, axis=1)
```

```python
data_subset.head()
```

```python
data_subset.isnull().sum()
```

```python
data_subset['Revenue (Millions)'].fillna(value=data_subset['Revenue (Millions)'].mean(), inplace=True)
data_subset['Metascore'].fillna(value=data_subset['Metascore'].mean(), inplace=True)
```

```python
data_subset.isnull().sum()
```

```python
genre=data_subset.Genre.str.split(',', expand=True).stack()
genre_dummies=pd.get_dummies(genre, prefix='g').groupby(level=0).sum()
```

```python
genre_dummies.head()
```

```python
genre_dummies.columns
```

```python
data_subset_2 = data_subset.drop(['Genre','Director'], axis=1)
data_dummies = pd.concat([genre_dummies,data_subset_2], axis=1)
```

```python
data_dummies.head()
```

```python
data_dummies.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler()
data_scaled=sc_X.fit_transform(data_dummies)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix= cosine_similarity(X=data_scaled)
```

```python
similarity_matrix.shape
```

```python
similarity_matrix[:10,:10]
```

```python
similarity_threshold=0.7
similarity = pd.DataFrame(similarity_matrix, index=data_dummies.index)
```

```python
similarity.head()
```

```python
movie_to_rec = list(np.where(similarity_matrix[0,:] >= 0.9))
movie_to_rec
```

```python
data.Title[data.index.isin((movie_to_rec))]
```

```python
list(movie_to_rec)
```

```python
movies_to_rec = []
for i in range(len(similarity_matrix)):
    if similarity_matrix[1,i] >= 0.8:
        movies_to_rec.append(data[data.index == i].Title.values)
print(movies_to_rec)
```
